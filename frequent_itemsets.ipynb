```python
import numpy as np
```

```python
n = int(input('Enter the number of transactions: '))
transactions = []
for i in range(n):
    transactions.append(list(map(int, input(f'Enter the items in transaction {i+1}:').split(','))))
```

```python
transactions
```

```python
itemset = []
flattened_list = [item for sublist in transactions for item in sublist]
elements = set(flattened_list)
for i in elements:
    count = 0 
    for transaction in transactions:
        if i in transaction:
            count += 1
    if count >= 2:
        itemset.append(i)
```

```python
itemset
```

```python
from itertools import combinations

subsets = []
for i in range(len(itemset) + 1):
    subsets.extend(combinations(itemset, i))
```

```python
subsets
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori

import warnings
warnings.filterwarnings('ignore')
```

```python
n = int(input('Enter the number of transactions: '))
transactions = []
for i in range(n):
    transactions.append(list(map(int, input(f'Enter the items in transaction {i+1}:').split(','))))
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support = 0.4, use_colnames=True)  
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

association_rules_df = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6) 
```
