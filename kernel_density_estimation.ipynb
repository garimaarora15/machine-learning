```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_kernel(x, data, h):
    n = len(data)
    pdf = np.zeros_like(x)
    for xi in data:
        pdf += 1/(np.sqrt(2*np.pi)*h) * np.exp(-(x-xi)**2 / (2*h**2))
    return pdf

def plot_gaussian_kernel(data, h):
    x = np.linspace(min(data) - 1, max(data) + 1, 1000)
    pdf = gaussian_kernel(x, data, h)
    plt.plot(x, pdf, label='Gaussian kernel pdf')
    plt.xlabel('data')
    plt.ylabel('density')
    plt.title("Gaussian kernel density")
    plt.show()

h = float(input('Enter the value of h: '))
data = list(map(float, input('Enter the value of data points: ').split(',')))
plot_gaussian_kernel(data, h)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def uniform_kernel(x, data, h):
    n = len(data)
    pdf = np.zeros_like(x)
    for xi in data:
        pdf += (1 / h) * np.where(np.abs(x - xi) <= h/2, 1, 0)
    pdf /= (n * h)  # Adjusting by the total area under the kernel
    return pdf

def plot_uniform_kernel(data, h):
    x = np.linspace(min(data) - 1, max(data) + 1, 1000)
    pdf = uniform_kernel(x, data, h)
    plt.plot(x, pdf, label='Uniform kernel pdf')
    plt.xlabel('data')
    plt.ylabel('density')
    plt.title("Uniform kernel density")
    plt.show()

h = float(input('Enter the value of h: '))
data = list(map(float, input('Enter the value of data points: ').split(',')))
plot_uniform_kernel(data, h)
```
