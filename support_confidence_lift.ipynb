```python
import pandas as pd
import numpy as np

n = int(input('Enter the number of transactions: '))
transactions = []
for i in range(n):
    transactions.append(list(map(str, input(f'Enter the items in transaction {i+1}:').split(','))))

def calculate_metrics(antecedent, consequent, transactions):
    count  = count_antecedent = count_consequent = 0
    for transaction in transactions:
        if set(antecedent).issubset(transaction):
            count_antecedent += 1
        if set(consequent).issubset(transaction):
            count_consequent += 1
        if set(antecedent).issubset(transaction) and set(consequent).issubset(transaction):
            count += 1
    support_antecedent = float(count_antecedent)/n
    support_consequent = float(count_consequent)/n
    support = float(count)/n
    confidence = float(count)/count_antecedent
    lift = support/(support_antecedent*support_consequent)
    return (support, confidence, lift)

ant = ['A']
cons = ['D']
calculate_metrics(ant, cons, transactions)
```
